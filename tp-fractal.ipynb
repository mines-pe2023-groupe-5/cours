## *La série de Mandelbrot*

La série de Mandelbrot, nommée d'après le mathématicien Benoît Mandelbrot, est un ensemble mathématique dont le résultat est une fractale aux formes complexes et fascinantes. Cet ensemble est défini par une suite récursive simple : \( Z_{n+1} = Z_n^2 + C \), où \( Z \) est un nombre complexe et \( C \) est une constante. La série est généralement visualisée sur le plan complexe, où les points qui n'appartiennent pas à l'ensemble de Mandelbrot sont colorés en fonction du nombre d'itérations nécessaires pour qu'ils s'échappent vers l'infini. Ces visualisations révèlent des structures détaillées et auto-similaires à toutes les échelles, faisant de la série de Mandelbrot un sujet d'étude passionnant en mathématiques et en informatique.

Plus de details -> https://fr.wikipedia.org/wiki/Ensemble_de_Mandelbrot
