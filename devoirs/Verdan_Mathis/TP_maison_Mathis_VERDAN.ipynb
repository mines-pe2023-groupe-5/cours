# Analyse des données

## Production des différents pays dans le monde

On va d'abord s'intéresser à la production totale des différents pays du monde

Créons pour cela une nouvelle colonne "Total"

```python
culture_capture["total_T"] = culture_capture["capture_T"]+culture_capture["culture_T"]
culture_capture.head()
```

Essayons de repérer les principaux pays producteurs de poissons (on regarde les données récentes)

```python
culture_capture[culture_capture.Year == 2018].sort_values('total_T', ascending = False)
```

Les cases vides m'embêtent, je vais les mettre à 0 faute d'informations

```python
culture_capture.fillna(0, inplace = True)
```

```python
culture_capture[culture_capture.Year == 2018].sort_values('total_T', ascending = False).head(15)
```

On remarque que certains pays sont regroupés en régions plus larges (pays riches, Europe ...) mais cela ne nous intéresse pas (on ne veut que les pays). Or ces régions correspondent aux lignes où la case Code est à 0.
(Je vais cette fois utiliser query car c'est plus simple) On récupère dans la suite les 15 premiers pays.

```python
table_1 = culture_capture.query("Code != 0 and Year == 2018").sort_values('total_T', ascending = False).head(15)
table_1 = table_1.drop(columns = ["culture_T", "capture_T", "Code", "Year"])
table_1
```

On va essayer de regrouper tous les autres pays ensembles.
La commande suivante permet de regrouper les autres pays et de calculer la production totale de tous ces pays avec la fonction sum(). On va ensuite rajouter une nouvelle ligne à table_1 à partir de ces données

```python
s = culture_capture.query("Code != 0 and Year == 2018").sort_values('total_T', ascending = False).iloc[15:,].drop(columns = ["culture_T", "capture_T"]).sum()
```

```python
s['total_T']
```

```python
row_other = ({"Entity" : ["Other"], "total_T" : [s["total_T"]]})
```

```python
pd.DataFrame(row_other)
```

On veut ensuite fusionner ce dataframe avec table_1 pour avoir un seul dataframe avec les 15 premiers pays puis tous les autres regroupés dans une ligne. Mais comme je ne sais pas me servir de .concat, j'utilise help() pour ouvrir la documentation.

```python
help(pd.concat([table_1, pd.DataFrame(row_other)]).plot)
```

On voit donc qu'au niveau mondial, la Chine et l'Indonésie pêchent la moitié des poisson du monde. Le fait que l'indonésie ait une activité de pêche aussi important peut s'expliquer par le fait que la ZEE de l'indonésie soit la 6ème mondiale et aussi car l'alimentation indonésienne comporte beaucoup plus de poisson que la notre par exemple.

## Production en culture vs capture

On va essayer de tracer le graphe de l'évolution de la production mondiale de poisson en culture vs capture

```python
table_2 = culture_capture.query('Code != 0')
table_2 = table_2.drop(columns = ['total_T'])
# on ne garde que les pays et on peut enlever le total, inutile ici
```

```python
table_2.head()
```

```python
table_2.sort_values('Year').Year.unique()
```

```python
print(table_2.query('Year == 1972').sum().culture_T)
```

Avec la ligne ci-dessus on peut récupérer la production de poisson en culture sur une année dans le monde entier, il reste à faire cela pour toutes les années entre 1960 et 2018 pour la culture et la capture. On va pour cela utiliser du 'vrai' Python.

```python
evo_cc = pd.DataFrame({'Year' : [] , 'Culture' : [], 'Capture' : []})

evo_cc
```

```python
for i in range(1960, 2016):
    evo_cc.loc[len(evo_cc)] = [i,
                             table_2.query('Year == {toto}'.format(toto = i)).sum().culture_T,
                             table_2.query('Year == {toto}'.format(toto = i)).sum().capture_T]
evo_cc.head(20)
```

```python
evo_cc.set_index('Year').plot(title='Production par capture vs culture dans le monde')
```

On retrouve le même graphique que sur le site internet, ce qui est rassurant

On peut de plus se dire que la surexploitation des poissons sauvages explique le fait que les captures de poissons stagnent depuis quelques dizaines d'années alors que l'aquaculture est de plus en plus importante (les réserves naturelles de poissons s'amenuisent et se font remplacer par de l'élevage).

## Evolution pays par pays

On va maintenant créer une fonction qui prend en paramètre le nom d'un pays et qui affiche l'évolution de la production de poisson

```python
def evo_pays(nom_pays) : 
    table = culture_capture.query(f"Entity == '{nom_pays}'")
    table = table.drop(columns = ['Entity', 'Code'])
    table = table.set_index('Year')
    table.plot(title=f"Evolution pour le pays {nom_pays}")
```

```python
evo_pays('China')
```

```python
evo_pays('France')
```

On voit clairement pourquoi la Chine est dite 'pays en développement' par rapport à la France : la production de la France est stable depuis 60 ans alors que celle de la Chine a explosé aux alentours de 1995.

## Evolution de chaque pays dans le monde

Enfin, on va voir la part que produit chaque pays dans le monde (en gros on reprend la première étape mais sur plusieurs années)

```python
evo_monde = culture_capture.query("Code != 0")
evo_monde.head()
```

```python
evo_monde.sort_values('total_T', ascending=False).head(10)
```

Le 'world' (qui n'apparait pas en 2018) nous embête pour les autres années, donc on le supprime

```python
evo_monde = evo_monde.query("Entity != 'World' ")
```

```python
evo_monde.sort_values('total_T', ascending=False).head(10)
```

On va tracer  sur un même graphique l'évolution de la production des plus gros producteurs dans le monde. Pour cela, on va regrouper les données par pays.

```python
gp_pays = evo_monde.groupby(by = 'Entity')
```

```python
gros_pays = gp_pays.total_T.max().sort_values().tail(10).index
```

Cette commande donne la liste des plus gros producteurs

```python
import matplotlib.pyplot as plt
```

```python
for pays in gros_pays : 
    truc = evo_monde.query(f"Entity == '{pays}'")
    truc = truc.set_index('Year')
    plt.plot(truc.index, truc.total_T, label=f"{pays}")
plt.title("Evolution de la production de poisson des principaux pays du monde.")
plt.legend()
plt.show()
```

On voit donc que les Chinois pêchent beaucoup de poisson et on retrouve les données du graphique tracé précédement à la date 2018.

## Conclusion

Cette analyse nous a montré que la situation économique d'un pays est reflétée par son activité de pêche (plus le pays se développe, plus il pêche du poisson, et les guerres font diminuer la production). Les pays industrialisés depuis longtemps ont une activité quasi constante à l'inverse de la Chine qui se développe très vite. Cela se voit car depuis 1995, la Chine surpasse tous les autres pays en terme de production.
Enfin, comme les poissons sauvages sont surexploités, la part de la pêche sauvage diminue, si bien que l'aquaculture devient majoritaire.
