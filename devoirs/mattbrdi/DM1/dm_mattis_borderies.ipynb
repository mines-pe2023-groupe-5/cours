```python
import pandas as pd
url = "base_données.xlsx"
df = pd.read_excel(url)
```

```python
df.head()
```

```python
#En NBA, les joueurs peuvent régulierement être Trade, c'est à dire échanger 
#chaque joueur apparait donc plusieurs fois si il a changé d'équipe pendant, trouvons le joueur avec le plus de points moyens
#Et on ne peut donc pas indexer par le nom du joueur 
df["total_point"] = df['G'] * df['PTS'] #on calcule tout d'abord le nombre total de point marqué
df.head()
#on remarque des nombre à virgules, surement un arrondi de la moyenne 
```

```python
#pour faire ca, on va grouper par nom de joueur, dans l'idée de voir qui à le plus marquer 
group = df.groupby("Player")
print(group.total_point.sum().idxmax()) #le joueur avec le nombre max de point marqué sur la saison
print(group.total_point.sum().max()) #le nombre max de point marqué sur la saison 
#mais cette donnée ne nous interessent finalement pas, il a joué beaucoup plus de match que les autres 
#on cherche donc le nombre de point par match en moyenne 
```

```python
new_df = {}
Nom = []
Point_par_match = []
for name, joueur in group: #on parcours le dataframe associé à chaque joueur
    data = group.get_group(name)
    point = data.total_point.sum()
    match = data.G.sum()
    ppg = point/match
    Nom.append(name)
    Point_par_match.append(ppg)

new_df ={'Player' : Nom,
        'ppg' : Point_par_match}
df1 = pd.DataFrame(new_df)
df1.set_index('Player')
print(df1.idxmax)
print(df1.max())



    
    
```
