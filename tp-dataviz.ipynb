# Visualisation

## Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
print(tips.dtypes)
tips.describe()
```

```python
tips.head()
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

## Graphs

```python
sns.relplot(tips, x="total_bill", y="tip", hue="time")
```

```python
tips["percent"] = tips.tip / tips.total_bill
```

```python
fig = sns.displot(data=tips, x="percent", kde=True)
fig
```

```python
sns.barplot?
```

```python
ax = sns.barplot(data=flight, x="year", y="passengers", hue="year", errorbar=None, estimator="sum")
ax.set(title="Passengers per year")
ax.set(ylabel=None)
```

```python
ax = sns.lineplot(data=flight, x="year", y="passengers", marker="o", estimator="sum", errorbar=None)
```
